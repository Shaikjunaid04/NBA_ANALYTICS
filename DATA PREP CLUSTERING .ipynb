```python
# Calculate composite score
df1_2018['Composite_Score'] = (
    df1_2018['PTS'] +
    df1_2018['W']+
    df1_2018['ORB']+
    df1_2018['DRB']+
    df1_2018['W/L%'] +
    df1_2018['FG%'] +
    df1_2018['3P%'] +
    df1_2018['2P%'] +
    df1_2018['eFG%'] )

# Rank players based on the composite score
df1_2018['Rank'] = df1_2018['Composite_Score'].rank(ascending=False, method='min')

# Select top 200 players
DF_2018 = df1_2018[df1_2018['Rank'] <= 50]


# Drop the temporary columns used for scoring and ranking
DF_2018 = DF_2018.drop(['Composite_Score', 'Rank'], axis=1)

# Display the resulting DataFrame
print(DF_2018)

```
