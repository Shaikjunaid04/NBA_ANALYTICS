```python
import pandas as pd
import numpy as np
df=pd.read_csv("CLEANED_FINAL_DATA_1.csv")
```

```python
df
```

```python

# Drop unnecessary attribute

df = df[['PTS', 'ORB', 'DRB', '3P%', '2P%', 'FG%','FT%']]

#age_bins = [20, 30, 40, np.inf]
#age_labels = ['Youngest', 'Middle Age', 'eldest', ]
#df['Age'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels)

# Example binning for 'PTS'
pts_bins = [0.001,10,20,30]
pts_labels = ['0-10 Low PTS', '10.1-20 Average PTS', '20.1-30 Highest PTS']
df['PTS'] = pd.cut(df['PTS'], bins=pts_bins, labels=pts_labels)

# Example binning for 'Wins'
#wins_bins = [0, 20, 40, 60, 80, 100]
#wins_labels = ['Less Wins', 'Average Wins', 'Good Wins', 'High Wins', 'Highest Wins']
#df['W'] = pd.cut(df['W'], bins=wins_bins, labels=wins_labels)

# Example binning for 'ORB' and 'DRB'
orb_bins = [0.01, 1, 2, 3, 4, 5]
orb_labels = ['0.1-1.0 ORB ', '1.0-2.0 ORB ', '2.0-3.0 ORB ', '3.0-4.0 ORB', '4.0 - 5.0 ORB']
df[['ORB']] = df[['ORB']].apply(lambda x: pd.cut(x, bins=orb_bins, labels=orb_labels))

drb_bins = [0.01, 1, 2, 3, 4, 5]
drb_labels = ['0.1-1.0 DRB', '1.0-2.0 DRB', '2.0-3.0 DRB', '3.0-4.0 DRB', '4.0 - 5.0 DRB']
df[['DRB']] = df[['DRB']].apply(lambda x: pd.cut(x, bins=drb_bins, labels=drb_labels))

# Example binning for '3P%', '2P%', 'FG%'
percentage_bins = [0.001, 0.2, 0.4, 0.6, 0.8, 0.999]
percentage_labels = ['0.0-0.19 3P%', '0.2-0.39 3P%', '0.4-0.59 3P%', '0.6-0.79 3P%', '0.8-0.999 3P%']
df[['3P%']] = df[['3P%']].apply(lambda x: pd.cut(x, bins=percentage_bins, labels=percentage_labels))

# Example binning for '3P%', '2P%', 'FG%'
percentage_bins = [0.001, 0.2, 0.4, 0.6, 0.8, 0.999]
percentage_labels = ['0.0-0.19 2P%', '0.2-0.39 2P%', '0.4-0.59 2P%', '0.6-0.79 2P%', '0.8-0.999 2P%']
df[['2P%']] = df[['2P%']].apply(lambda x: pd.cut(x, bins=percentage_bins, labels=percentage_labels))

# Example binning for '3P%', '2P%', 'FG%'
percentage_bins = [0.001, 0.2, 0.4, 0.6, 0.8, 0.999]
percentage_labels = ['0.0-0.19 FG%', '0.2-0.39 FG%', '0.4-0.59 FG%', '0.6-0.79 FG%', '0.8-0.999 FG%']
df[['FG%']] = df[['FG%']].apply(lambda x: pd.cut(x, bins=percentage_bins, labels=percentage_labels))

# Example binning for '3P%', '2P%', 'FG%'
percentage_bins = [0.001, 0.2, 0.4, 0.6, 0.8, 0.999]
percentage_labels = ['0.0-0.19 FT%', '0.2-0.39 FT%', '0.4-0.59 FT%', '0.6-0.79 FT%', '0.8-0.999 FT%']
df[['FT%']] = df[['FT%']].apply(lambda x: pd.cut(x, bins=percentage_bins, labels=percentage_labels))

```
